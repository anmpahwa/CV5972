
**Learning Objectives**

By the end of this notebook, you will be able to:
1. Construct the Lagrangian dual function and dual problem for a constrained NLP, and explain why the dual function is always concave.
2. State and apply the weak and strong duality theorems for convex NLPs, and recover LP duality (Lecture 9) as a special case.
3. Interpret KKT multipliers as dual variables and shadow prices, and verify strong duality by hand for transportation optimization problems.

**Prerequisites**: LP Duality (Lecture 9); NLP Principles (Lecture 12); KKT Conditions (Lectures 16–18); Penalty, Barrier, Interior-Point, and SQP Methods (Lectures 16–18).

**Estimated time**: 90 minutes (including in-class exercises)

---

## KKT Conditions

Consider the standard-form NLP:

$$\min_{x} \; f(x) \quad \text{subject to} \quad g_i(x) \leq 0, \quad i = 1, \ldots, m$$

where $f, g_i : \mathbb{R}^n \to \mathbb{R}$ are continuously differentiable. The **Lagrangian** is:

$$\mathcal{L}(x, \lambda) = f(x) + \sum_{i=1}^{m} \lambda_i \, g_i(x)$$

If $x^*$ is a local minimum and a constraint qualification holds (e.g., LICQ), then there exist multipliers $\lambda^* \in \mathbb{R}^m$ such that the following **Karush–Kuhn–Tucker (KKT) conditions** are satisfied:

| Condition | Statement | Interpretation |
|-----------|-----------|----------------|
| **Stationarity** | $\nabla f(x^*) + \sum_{i=1}^{m} \lambda_i^* \nabla g_i(x^*) = \mathbf{0}$ | Gradient of objective balanced by gradients of active constraints |
| **Primal feasibility** | $g_i(x^*) \leq 0 \quad \forall\, i$ | Solution lies within the feasible region |
| **Dual feasibility** | $\lambda_i^* \geq 0 \quad \forall\, i$ | Multipliers are non-negative for inequality constraints |
| **Complementary slackness** | $\lambda_i^* \, g_i(x^*) = 0 \quad \forall\, i$ | A constraint is either active ($g_i = 0$) or its multiplier is zero ($\lambda_i = 0$) |

For a **convex** NLP (convex $f$, convex $g_i$), the KKT conditions are both necessary and sufficient for global optimality. All methods in Lectures 16–18 seek a point satisfying these four conditions — and in doing so, they recover the multipliers $\lambda^*$. This lecture shows that $\lambda^*$ is precisely the optimal solution of the **Lagrangian dual problem**, and that $f(x^*) = q(\lambda^*)$ — strong duality.

---

## Constrained Optimization Algorithms — Overview

The NLP standard form (introduced in Lecture 12) is:

$$\min_{x} \; f(x) \quad \text{subject to} \quad g_i(x) \leq 0, \quad i = 1, \ldots, m$$

Lectures 16–18 introduced four algorithms that compute a KKT point. Each recovers the KKT multipliers $\lambda^*$ as a by-product:

| Method | Core idea | Feasibility of iterates | Penalty/barrier parameter | Convergence |
|--------|-----------|------------------------|--------------------------|-------------|
| **Penalty** (L16) | Append a penalty for constraint violation to $f$ | Iterates are **infeasible**; converge to feasibility as $\rho \to \infty$ | $\rho \to \infty$ (increased) | Exact KKT point recovered only in the limit |
| **Barrier** (L17) | Append a barrier that blows up near constraint boundary | Iterates are **strictly feasible**; approach boundary as $\mu \to 0$ | $\mu \to 0$ (decreased) | Exact KKT point recovered only in the limit |
| **Interior-Point** (L18) | Solve KKT system directly with barrier modification; Newton steps inside feasible region | Iterates are **strictly feasible** by construction | $\mu \to 0$ (decreased) | Polynomial-time; exact KKT point in the limit |
| **SQP** (L19) | Newton on KKT conditions; each step solves a linearized QP subproblem | Iterates may be **infeasible** until convergence | None | **Exact** for QP with linear constraints; **superlinear** in general |

```{note}
All four methods compute $\lambda^*$ as part of their output — but none has yet explained what $\lambda^*$ *means* beyond "the KKT multiplier." This lecture answers that question: $\lambda^*$ is the optimal solution of the Lagrangian dual, and its value has a precise shadow-price interpretation identical to the LP dual variables from Lecture 9.
```

## NLP Duality

### 1. The Lagrangian Dual Function

Fix a multiplier vector $\lambda \geq \mathbf{0}$ and minimize the Lagrangian over all $x \in \mathbb{R}^n$ — with no constraint on $x$ at all:

$$q(\lambda) = \inf_{x \in \mathbb{R}^n} \mathcal{L}(x, \lambda) = \inf_{x \in \mathbb{R}^n} \left[ f(x) + \sum_{i=1}^{m} \lambda_i \, g_i(x) \right]$$

The function $q : \mathbb{R}^m \to \mathbb{R} \cup \{-\infty\}$ is called the **Lagrangian dual function**. Three properties follow immediately:

**Property 1 — $q$ is always concave.** $q(\lambda)$ is the pointwise infimum of a family of affine functions of $\lambda$ (one for each fixed $x$). The infimum of affine functions is always concave, regardless of whether $f$ or $g_i$ are convex. This means the dual problem is always a concave maximization — and therefore has no local optima that are not global.

**Property 2 — $q$ lower-bounds the primal.** For any feasible $x$ (i.e., $g_i(x) \leq 0$ for all $i$) and any $\lambda \geq \mathbf{0}$:

$$q(\lambda) \leq \mathcal{L}(x, \lambda) = f(x) + \underbrace{\sum_{i=1}^{m} \lambda_i \, g_i(x)}_{\leq\, 0 \text{ (since } \lambda_i \geq 0,\, g_i(x) \leq 0)} \leq f(x)$$

So $q(\lambda) \leq f(x)$ for every feasible $x$ and every $\lambda \geq \mathbf{0}$. Taking the infimum over $x$ on the right gives: $q(\lambda) \leq f^*$ for all $\lambda \geq \mathbf{0}$, where $f^* = f(x^*)$ is the primal optimal value. Every dual-feasible $\lambda$ provides a **valid lower bound** on the primal optimal.

**Property 3 — When $q(\lambda) = -\infty$.** If for some direction in $\mathbb{R}^n$ the Lagrangian is unbounded below, $q(\lambda) = -\infty$. Dual-feasible multipliers are those for which $q(\lambda) > -\infty$; the set of such $\lambda$ is the **dual feasible region** $\mathcal{D} = \{\lambda \geq \mathbf{0} : q(\lambda) > -\infty\}$.

### 2. The Dual Problem

Since $q(\lambda)$ is a lower bound on $f^*$ for every $\lambda \in \mathcal{D}$, the tightest such bound is the **Lagrangian dual problem**:

$$\max_{\lambda \geq \mathbf{0}} \; q(\lambda)$$

Let $q^* = \max_{\lambda \geq \mathbf{0}} q(\lambda)$ denote the dual optimal value, attained at some $\lambda^*$. The comparison with Lecture 9:

| | LP (Lecture 9) | NLP (Lecture 20) |
|--|----------------|------------------|
| **Primal** | $\max_x \; \mathbf{c}^\top x$ s.t. $\mathbf{A}x \leq \mathbf{b}$, $x \geq 0$ | $\min_x \; f(x)$ s.t. $g_i(x) \leq 0$ |
| **Dual** | $\min_y \; \mathbf{b}^\top y$ s.t. $\mathbf{A}^\top y \geq \mathbf{c}$, $y \geq 0$ | $\max_{\lambda \geq 0} \; q(\lambda)$ |
| **Dual objective** | Linear in $y$ | $q(\lambda) = \inf_x \mathcal{L}(x,\lambda)$ — always concave |
| **Dual feasible region** | Polyhedral ($\mathbf{A}^\top y \geq \mathbf{c}$) | $\mathcal{D} = \{\lambda \geq 0 : q(\lambda) > -\infty\}$ |

### 3. Weak Duality

```{note}
**Weak Duality Theorem**: For any primal-feasible $x$ and any $\lambda \in \mathcal{D}$:

$$q(\lambda) \leq f^* \leq f(x)$$

In particular, the dual optimal value never exceeds the primal optimal value: $q^* \leq f^*$.
```

The gap $f^* - q^* \geq 0$ is the **duality gap**. Weak duality holds for *all* optimization problems — convex or not, linear or not — as long as $\lambda \geq \mathbf{0}$.

### 4. Strong Duality

For general NLPs, the duality gap may be strictly positive: the dual optimal cannot reach the primal optimal. For **convex** NLPs, however, the gap closes under a mild regularity condition:

```{note}
**Slater's Condition**: There exists a point $\hat{x} \in \mathbb{R}^n$ (a **Slater point**) such that $g_i(\hat{x}) < 0$ strictly for all $i = 1, \ldots, m$.

**Strong Duality Theorem (convex NLP)**: If the NLP is convex (convex $f$, convex $g_i$) and Slater's condition holds, then:
1. $q^* = f^*$ — the duality gap is zero.
2. The dual optimal $\lambda^*$ exists and satisfies the KKT conditions at $x^*$.
```

In other words, for convex NLPs satisfying Slater's condition, solving the dual is equivalent to solving the primal — and the dual optimal multipliers are exactly the KKT multipliers recovered by Penalty, Barrier, Interior-Point, and SQP in Lectures 16–18.

### 5. LP Duality as a Special Case

Substitute $f(x) = -\mathbf{c}^\top x$ (negated for minimization), $g_i(x) = \mathbf{A}_i x - b_i$ into the Lagrangian:

$$q(\lambda) = \inf_{x \geq 0} \left[ -\mathbf{c}^\top x + \lambda^\top (\mathbf{A}x - \mathbf{b}) \right] = -\mathbf{b}^\top \lambda + \inf_{x \geq 0} (\mathbf{A}^\top \lambda - \mathbf{c})^\top x$$

The infimum over $x \geq 0$ is $-\infty$ unless $(\mathbf{A}^\top \lambda - \mathbf{c}) \geq \mathbf{0}$ (otherwise one can send $x_j \to \infty$). When the constraint holds, the infimum equals zero. So:

$$q(\lambda) = \begin{cases} -\mathbf{b}^\top \lambda & \text{if } \mathbf{A}^\top \lambda \geq \mathbf{c} \\ -\infty & \text{otherwise} \end{cases}$$

The dual problem $\max_{\lambda \geq 0} q(\lambda)$ then becomes:

$$\max_{\lambda \geq 0} \; (-\mathbf{b}^\top \lambda) \quad \text{s.t.} \quad \mathbf{A}^\top \lambda \geq \mathbf{c}$$

Renaming $\lambda \to y$ and flipping the sign of the objective to match the LP convention:

$$\min_{y \geq 0} \; \mathbf{b}^\top y \quad \text{s.t.} \quad \mathbf{A}^\top y \geq \mathbf{c}$$

This is exactly the LP dual from Lecture 9. LP always satisfies Slater's condition when the primal is feasible and bounded (any interior feasible point is a Slater point), so strong duality holds without further assumption — recovering the LP Strong Duality Theorem as a corollary of the NLP result.

### 6. Shadow-Price Interpretation of $\lambda^*$

Strong duality makes the economic interpretation of $\lambda_i^*$ precise. Define the **perturbed primal** by shifting constraint $i$ by a small amount $\epsilon$:

$$f^*(\epsilon) = \min_{x} f(x) \quad \text{s.t.} \quad g_i(x) \leq \epsilon, \quad g_j(x) \leq 0 \; (j \neq i)$$

Under mild regularity conditions (convexity, Slater):

$$\lambda_i^* = -\frac{\partial f^*(\epsilon)}{\partial \epsilon}\bigg|_{\epsilon=0}$$

That is, $\lambda_i^*$ measures how much the optimal cost *decreases* per unit of *relaxation* in constraint $i$. Tightening the constraint (negative $\epsilon$) increases cost at rate $\lambda_i^*$; relaxing it (positive $\epsilon$) decreases cost at the same rate.

| KKT condition | Dual interpretation |
|---|---|
| $\lambda_i^* \geq 0$ | Relaxing any constraint can only help (or not affect) the objective — costs never *increase* when you loosen a budget |
| $\lambda_i^* g_i(x^*) = 0$ (complementary slackness) | An inactive constraint ($g_i < 0$) is not binding, so relaxing it further changes nothing: $\lambda_i^* = 0$ |
| $\lambda_i^* > 0$ | Constraint $i$ is active and binding — the resource is scarce and carries a positive shadow price |

This is identical to the shadow-price interpretation of LP dual variables from Lecture 9 — generalized to non-linear cost and constraint functions.

## In-Class Exercises

### Exercise 1

*Same problem as Lectures 16–18 In-Class Exercise 1*

MTC Chennai is optimizing a demand-responsive feeder service on the Tambaram–Chromepet corridor. The daily operating cost depends on fleet size $n$ (vehicles) and scheduled headway $h$ (minutes):

$$C(n, h) = (n - 5)^2 + 2(h - 3)^2$$

The depot has a combined resource constraint — vehicles and crew shifts — that limits $n + h \leq 7$.

From Lectures 16–18, the constrained optimum is $\left(n^* = \tfrac{13}{3},\, h^* = \tfrac{8}{3}\right)$ with KKT multiplier $\lambda^* = \tfrac{4}{3}$.

**Part 1 — Construct the Lagrangian and dual function.**

The constraint in standard form is $g(n, h) = n + h - 7 \leq 0$. Form the Lagrangian:

$$\mathcal{L}(n, h, \lambda) = (n-5)^2 + 2(h-3)^2 + \lambda(n + h - 7)$$

To compute $q(\lambda) = \inf_{n, h} \mathcal{L}(n, h, \lambda)$, treat $\lambda \geq 0$ as fixed and minimize over $(n, h)$ without any constraint:

$$\frac{\partial \mathcal{L}}{\partial n} = 2(n - 5) + \lambda = 0 \implies n^*(\lambda) = 5 - \frac{\lambda}{2}$$

$$\frac{\partial \mathcal{L}}{\partial h} = 4(h - 3) + \lambda = 0 \implies h^*(\lambda) = 3 - \frac{\lambda}{4}$$

Substituting back into $\mathcal{L}$:

$$q(\lambda) = \left(-\frac{\lambda}{2}\right)^2 + 2\left(-\frac{\lambda}{4}\right)^2 + \lambda\left(\left(5 - \frac{\lambda}{2}\right) + \left(3 - \frac{\lambda}{4}\right) - 7\right)$$

$$= \frac{\lambda^2}{4} + \frac{\lambda^2}{8} + \lambda\left(1 - \frac{3\lambda}{4}\right) = \lambda - \frac{3\lambda^2}{8}$$

This is a concave quadratic in $\lambda$ — confirming Property 1 (the dual function is always concave).

**Part 2 — Solve the dual problem.**

$$\max_{\lambda \geq 0} \; q(\lambda) = \lambda - \frac{3\lambda^2}{8}$$

Taking the derivative and setting it to zero:

$$\frac{dq}{d\lambda} = 1 - \frac{3\lambda}{4} = 0 \implies \lambda^* = \frac{4}{3}$$

Since $q$ is concave, this is the global maximum. The dual optimal value is:

$$q\!\left(\tfrac{4}{3}\right) = \frac{4}{3} - \frac{3}{8}\left(\frac{4}{3}\right)^2 = \frac{4}{3} - \frac{3}{8} \cdot \frac{16}{9} = \frac{4}{3} - \frac{2}{3} = \frac{2}{3}$$

**Part 3 — Verify strong duality.**

The primal optimal cost is:

$$C\!\left(\tfrac{13}{3}, \tfrac{8}{3}\right) = \left(\tfrac{13}{3} - 5\right)^2 + 2\left(\tfrac{8}{3} - 3\right)^2 = \left(-\tfrac{2}{3}\right)^2 + 2\left(-\tfrac{1}{3}\right)^2 = \frac{4}{9} + \frac{2}{9} = \frac{2}{3}$$

$$\boxed{f^* = q^* = \tfrac{2}{3}} \qquad \checkmark \quad \text{(zero duality gap)}$$

Slater's condition is satisfied: the point $(n, h) = (1, 1)$ has $n + h = 2 < 7$, so it is strictly feasible.

**Part 4 — Verify complementary slackness.**

The constraint is active at the optimum: $n^* + h^* = \tfrac{13}{3} + \tfrac{8}{3} = 7$, so $g(n^*, h^*) = 0$. Complementary slackness requires $\lambda^* \cdot g(n^*, h^*) = \tfrac{4}{3} \cdot 0 = 0$ ✓.

**Part 5 — Interpret $\lambda^* = \tfrac{4}{3}$.**

If the combined resource budget tightens from 7 to 6 (i.e., $\epsilon = -1$), the optimal cost increases by approximately $\lambda^* \cdot 1 = \tfrac{4}{3}$ — an additional ₹1,333 per day in operating cost. Conversely, relaxing the budget from 7 to 8 saves approximately ₹1,333 per day. The resource is currently binding and carries a positive scarcity price.

### Exercise 2

*Same problem as Lectures 16–18 In-Class Exercise 2*

An NHAI corridor study models total system travel time across two parallel links — the Mumbai–Pune Expressway ($v_1$, thousand veh/hr) and the old NH48 ($v_2$, thousand veh/hr):

$$T(v_1, v_2) = 3v_1^2 - 2v_1 v_2 + 2v_2^2 - 14v_1 - 12v_2 + 50$$

A combined capacity constraint limits total flow: $v_1 + v_2 \leq 5$.

From Lectures 16–18, the constrained optimum is $\left(v_1^* = \tfrac{16}{7},\, v_2^* = \tfrac{19}{7}\right)$ with KKT multiplier $\lambda^* = \tfrac{40}{7}$.

**Part 1 — Construct the Lagrangian and dual function.**

Constraint in standard form: $g(v_1, v_2) = v_1 + v_2 - 5 \leq 0$. Lagrangian:

$$\mathcal{L}(v_1, v_2, \lambda) = 3v_1^2 - 2v_1 v_2 + 2v_2^2 - 14v_1 - 12v_2 + 50 + \lambda(v_1 + v_2 - 5)$$

Minimizing over $(v_1, v_2)$ for fixed $\lambda \geq 0$:

$$\frac{\partial \mathcal{L}}{\partial v_1} = 6v_1 - 2v_2 - 14 + \lambda = 0$$

$$\frac{\partial \mathcal{L}}{\partial v_2} = -2v_1 + 4v_2 - 12 + \lambda = 0$$

Solving this $2 \times 2$ linear system (multiply first equation by 2, add to second):

$$10v_1 - 40 + 3\lambda = 0 \implies v_1^*(\lambda) = 4 - \frac{3\lambda}{10}$$

$$v_2^*(\lambda) = \frac{12 - \lambda - 2v_1^*(\lambda)}{4} \cdot \frac{4}{4} = 5 - \frac{2\lambda}{5}$$

Substituting back, the dual function is:

$$q(\lambda) = 4\lambda - \frac{7\lambda^2}{20} - 8$$

This is a concave quadratic — confirming the dual function is always concave.

**Part 2 — Solve the dual problem.**

$$\frac{dq}{d\lambda} = 4 - \frac{7\lambda}{10} = 0 \implies \lambda^* = \frac{40}{7}$$

$$q\!\left(\tfrac{40}{7}\right) = 4 \cdot \frac{40}{7} - \frac{7}{20} \cdot \left(\frac{40}{7}\right)^2 - 8 = \frac{160}{7} - \frac{160}{7} \cdot \frac{1}{2} - 8 = \frac{80}{7} - 8 = \frac{24}{7}$$

**Part 3 — Verify strong duality.**

The primal optimal value is:

$$T\!\left(\tfrac{16}{7}, \tfrac{19}{7}\right) = 3\left(\tfrac{16}{7}\right)^2 - 2 \cdot \tfrac{16}{7} \cdot \tfrac{19}{7} + 2\left(\tfrac{19}{7}\right)^2 - 14 \cdot \tfrac{16}{7} - 12 \cdot \tfrac{19}{7} + 50 = \frac{24}{7}$$

$$\boxed{f^* = q^* = \tfrac{24}{7}} \qquad \checkmark$$

**Part 4 — Interpret $\lambda^* = \tfrac{40}{7} \approx 5.71$.**

If the combined corridor capacity were reduced from 5,000 to 4,000 vehicles/hour (a tightening of 1 unit), total system travel time would increase by approximately $\tfrac{40}{7}$ thousand vehicle-hours — a substantial congestion cost. Conversely, a capacity expansion by 1,000 veh/hr (e.g., adding a third lane) saves approximately $\tfrac{40}{7}$ thousand vehicle-hours. NHAI can use this shadow price to evaluate whether the annualized cost of expansion is justified by the congestion savings.
