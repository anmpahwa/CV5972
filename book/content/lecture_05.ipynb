## Why Use a Spreadsheet Solver?

In Lecture 4 we solved LPs by hand — the graphical method for 2-variable problems and the Simplex algorithm for small instances. Real transportation problems, however, involve tens or hundreds of variables and constraints. Manually pivoting through a tableau becomes impractical.

The **Excel Solver** is a practical, widely-used tool that:
- Requires no programming — ideal for quick, one-off analyses
- Is used extensively in industry: logistics planners, fleet managers, and traffic engineers regularly use it for day-to-day decisions

> **Real-world hook**: A freight company like Delhivery or Blue Dart might use an Excel Solver model every morning to decide how many tonnes to allocate to road vs. rail, or which vehicles to assign to which corridors — before dispatching their fleet.

In this lecture we will solve*two problems in class and you will practise three problems as take-away exercises.

## The Excel Solver Workflow

Setting up any LP in Excel follows the same five-step template:

| Step | What you do in Excel |
|------|----------------------|
| **1. Data section** | Enter all problem parameters (costs, capacities, demands) in clearly labelled cells |
| **2. Decision variables** | Reserve a row/column of cells for $x_1, x_2, \ldots$ — start with value 0 |
| **3. Objective cell** | Write a formula: `=SUMPRODUCT(coefficients, decision_vars)` |
| **4. Constraint cells** | Write LHS formulas for each constraint; enter RHS values separately |
| **5. Solver dialog** | Set objective → set variable cells → add constraints → choose *Simplex LP* → Solve |

### Solver Dialog Settings

- **Set Objective**: the cell containing your objective formula
- **To**: Min or Max
- **By Changing Variable Cells**: the range holding decision variables
- **Subject to the Constraints**: click *Add* for each constraint
- **Select a Solving Method**: always choose **Simplex LP** for linear programmes
- After solving: select **"Keep Solver Solution"**

## In-Class Exercises 

### Exercise 1: Freight Mode Allocation

#### Problem Statement

A logistics company operates a Chennai–Delhi freight corridor. It can ship cargo either by truck or by rail. The planning horizon is one week.

| | Truck | Rail |
|--|-------|------|
| Revenue (₹/tonne) | 4,500 | 3,200 |
| Operating cost (₹/tonne) | 2,800 | 1,400 |
| CO₂ emissions (kg/tonne) | 62 | 18 |

Constraints:

1. Total capacity: at most 500 tonnes can be shipped in a week (combined)
2. Rail capacity: at most 300 tonnes by rail (track allocation limit)
3. Minimum service: at least 100 tonnes must go by truck (contractual obligation)
4. Emission cap: total CO₂ must not exceed 24,000 kg
5. Non-negativity: tonnes shipped ≥ 0

Maximise total profit (revenue − operating cost).

---

#### Step 1 — Formulate the LP

Let:
- $x_T$ = tonnes shipped by truck
- $x_R$ = tonnes shipped by rail

Objective:

$$
\max \ Z = (4500 - 2800)\,x_T + (3200 - 1400)\,x_R = 1700\,x_T + 1800\,x_R
$$

Subject to:

$$
\begin{aligned}
x_T + x_R \leq 500 & & \text{(Capacity)} \\
x_R \leq 300 & & \text{(Rail cap)} \\
x_T \geq 100 & & \text{(Min truck)} \\
62\,x_T + 18\,x_R \leq 24000 & & \text{(Emission cap)} \\
x_T,\, x_R \geq 0 & &
\end{aligned}
$$

---

#### Step 2 — Excel Spreadsheet Layout

Set up your spreadsheet exactly as follows:

```
      A                    B          C
 1    FREIGHT MODE ALLOCATION MODEL
 2
 3    Parameter            Truck      Rail
 4    Profit (₹/tonne)     1700       1800
 5    CO2 (kg/tonne)       62         18
 6
 7    Decision Variables   Truck      Rail
 8    Tonnes shipped       0          0         ← Solver changes these
 9
10    Objective
11    Total Profit (₹)     =SUMPRODUCT(B4:C4, B8:C8)
12
13    Constraints          LHS        RHS        Slack
14    Total capacity       =SUM(B8:C8)    500    =C14-B14
15    Rail capacity        =C8            300    =C15-B15
16    Min truck            =B8            100    =B16-C16
17    Emission cap         =SUMPRODUCT(B5:C5,B8:C8)  24000  =C17-B17
```

#### Step 3 — Solver Configuration

- **Set Objective**: B11 → **Max**
- **Variable Cells**: B8:C8
- **Constraints**:
  - B14 ≤ C14
  - B15 ≤ C15
  - B16 ≥ C16
  - B17 ≤ C17
  - B8:C8 ≥ 0
- **Method**: Simplex LP → Solve

---

#### Step 4 — Expected Results & Interpretation

After solving, verify:

| Variable | Value |
|----------|-------|
| $x_T$ (truck) | 200 tonnes |
| $x_R$ (rail) | 300 tonnes |
| **Total Profit** | ₹8,80,000 |

> **Managerial insight**: Both the total capacity constraint and the rail capacity constraint are binding at optimum — every available tonne of capacity is being used, and the rail allocation is fully exhausted. The minimum truck obligation and emission cap are both slack: the operator is already exceeding the minimum truck requirement, and emissions are well within the cap, so neither constraint is a binding concern at current demand levels.

### Exercise 2

#### Problem Statement

A signalised intersection in Kolkata has 4 phases. The traffic engineer wants to allocate green time (in seconds) across phases to maximise total vehicle throughput (vehicles/cycle).

| Phase | Movement | Saturation flow (veh/s) | Min green (s) | Max green (s) |
|-------|----------|------------------------|---------------|---------------|
| 1 | Northbound through | 0.52 | 10 | 45 |
| 2 | Southbound through | 0.48 | 10 | 45 |
| 3 | Eastbound through  | 0.35 | 8  | 35 |
| 4 | Westbound through  | 0.30 | 8  | 35 |

Constraints:

1. Cycle length = 90 seconds (fixed): $g_1 + g_2 + g_3 + g_4 = 90$
   *(we subtract a fixed lost time of 2s/phase → effective green budget = 90 − 4×2 = 82s)*
2. Minimum green: $g_i \geq g_i^{\min}$ for each phase $i$
3. Maximum green: $g_i \leq g_i^{\max}$ for each phase $i$

Maximise total throughput $= \sum_i s_i \cdot g_i$ where $s_i$ is saturation flow.

---

#### Step 1 — LP Formulation

Let $g_i$ = effective green time (seconds) allocated to phase $i$.

Objective:

$$
\max \ Z = 0.52\,g_1 + 0.48\,g_2 + 0.35\,g_3 + 0.30\,g_4
$$

Subject to:

$$
\begin{aligned}
g_1 + g_2 + g_3 + g_4 = 82 & & \text{(Cycle budget)} \\
g_i \geq g_i^{\min}, \quad i = 1,2,3,4 & & \text{(Min green)} \\
g_i \leq g_i^{\max}, \quad i = 1,2,3,4 & & \text{(Max green)} \\
\end{aligned}
$$

> **Note**: This is an LP with one **equality** constraint. In Solver, enter it as `LHS = RHS` (use the `=` option in the Add Constraint dialog).

---

#### Step 2 — Excel Spreadsheet Layout

```
      A                  B      C      D      E
 1    SIGNAL TIMING OPTIMISATION
 2
 3    Parameter          Ph 1   Ph 2   Ph 3   Ph 4
 4    Sat. flow (veh/s)  0.52   0.48   0.35   0.30
 5    Min green (s)      10     10     8      8
 6    Max green (s)      45     45     35     35
 7
 8    Decision Vars      Ph 1   Ph 2   Ph 3   Ph 4
 9    Green time (s)     0      0      0      0    ← Solver changes
10
11    Objective
12    Throughput (veh)   =SUMPRODUCT(B4:E4, B9:E9)
13
14    Constraints        LHS    RHS
15    Cycle budget       =SUM(B9:E9)    82     ← equality
16    Min Ph1            =B9    10
17    Min Ph2            =C9    10
18    Min Ph3            =D9    8
19    Min Ph4            =E9    8
20    Max Ph1            =B9    45
21    Max Ph2            =C9    45
22    Max Ph3            =D9    35
23    Max Ph4            =E9    35
```

#### Step 3 — Solver Configuration

- **Set Objective**: B12 → **Max**
- **Variable Cells**: B9:E9
- **Constraints**: B15 = C15 (equality!), then min/max bounds
- **Method**: Simplex LP → Solve

---

#### Step 4 — Expected Results & Interpretation

| Phase | Optimal green (s) |
|-------|-------------------|
| $g_1$ | 45 (at upper bound) |
| $g_2$ | 21 |
| $g_3$ | 8 (at lower bound) |
| $g_4$ | 8 (at lower bound) |
| **Throughput** | ≈ 38.7 veh/cycle |

> **Interpretation**: The solver assigns maximum green to Phase 1 (highest saturation flow) and Phase 2 gets the remaining budget after minimums are satisfied for Phases 3 and 4.

| | Standard Bus | Articulated Bus |
|--|-------------|-----------------|

MTC must satisfy five operational constraints. On the demand side, Route A must deliver at least 3,200 pass-km/day and Route B at least 1,800 pass-km/day. On the supply side, the daily operating budget cannot exceed ₹2,00,000, the driver pool is limited to 60, and the total fleet to 25 buses.

Assume each bus type can be split between routes. Let:
- $x_1$ = standard buses deployed (total across both routes)
- $x_2$ = articulated buses deployed (total across both routes)

Minimise total daily cost.
