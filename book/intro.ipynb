# CV5972 - Optimization Techniques for Transportation Engineering

### Objectives

- To develop fundamental understanding of optimization theory and techniques pertaining to linear and non-linear programming.
- To formulate and solve optimization problems in transportation engineering using suitable methods.
- To interpret and evaluate solutions derived from optimization models and support informed decision-making in real-world transportation scenarios.

### Content

- Module 1. Linear Programming
  - principles of linear programming
  - methods (graphical method, spreadhseet-based solvers, and simplex algorithm)
  - managerial decision-making with sensitivity analysis and duality
  - applications on variants of maximum-flow problem, least-cost path problem, transportation problem, and trans-shipment problem, among others.

- Module 2. Non-Linear Programming - Exact Methods
	- principles of non-linear programming 
	- methods (Karush-Kuhn-Tucker conditions, sequential quadratic programming, and interior-point methods)
  - managerial decision-making with duality
  - applications on variants of facility location problem and traffic assignment problem, among others.

- Module 3. Non-Linear Programming - Approximate Methods 
	- fundamentals of heuristics 
	- methods (local search heuristics, evolutionary computation techniques, and swarm intelligence algorithms)
  - managerial decision-making 
  - applications on variants of traveling salesman problem and vehicle routing problem, among others.
