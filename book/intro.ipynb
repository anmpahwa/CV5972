# CV5972 - Optimisation Techniques for Transportation Engineering

### Objectives

- To develop fundamental understanding of optimisation theory and techniques pertaining to linear and non-linear programming.
- To formulate and solve optimisation problems in transportation engineering using suitable methods.
- To interpret and evaluate solutions derived from optimisation models and support informed decision-making in real-world transportation scenarios.

### Content

- Module 1. Linear Programming
  - principles of linear programming
  - graphical method and simplex algorithm
  - sensitivity analysis and duality
  - inferring results and making decisions
  - applications on variants of transportation problem, assignment problem, trans-shipment problem, and least-cost path problem

- Module 2. Non-linear Programming
	- principles of non-linear programming 
	- exact methods (Karush-Kuhn-Tucker conditions, sequential quadratic programming, and interior-point methods)
  - non-exact methods (local search heuristics, evolutionary computation techniques, and swarm intelligence algorithms)
  - applications on variants of traveling salesman problem, vehicle routing problem, facility location problem, and network design problem
